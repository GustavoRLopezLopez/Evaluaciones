```python
import requests
import sqlite3

# Leer el JSON
url = "https://jsonplaceholder.typicode.com/users"
users = requests.get(url).json()

# Conexión a SQLite
conn = sqlite3.connect("../data/usuarios.db")
cursor = conn.cursor()
```

```python
# Tabla principal de usuarios
cursor.execute("""
CREATE TABLE IF NOT EXISTS usuarios (
    id INTEGER PRIMARY KEY,
    name TEXT,
    username TEXT,
    email TEXT,
    phone TEXT,
    website TEXT
)
""")

# Tabla de direcciones
cursor.execute("""
CREATE TABLE IF NOT EXISTS direcciones (
    user_id INTEGER,
    street TEXT,
    suite TEXT,
    city TEXT,
    zipcode TEXT,
    lat REAL,
    lng REAL,
    FOREIGN KEY(user_id) REFERENCES usuarios(id)
)
""")

# Tabla de empresas
cursor.execute("""
CREATE TABLE IF NOT EXISTS empresas (
    user_id INTEGER,
    name TEXT,
    catchPhrase TEXT,
    bs TEXT,
    FOREIGN KEY(user_id) REFERENCES usuarios(id)
)
""")
```

```python
for u in users:
    # Insertar en usuarios
    cursor.execute("""
    INSERT INTO usuarios (id, name, username, email, phone, website)
    VALUES (?, ?, ?, ?, ?, ?)
    """, (u['id'], u['name'], u['username'], u['email'], u['phone'], u['website']))
    
    # Insertar en direcciones
    addr = u['address']
    geo = addr['geo']
    cursor.execute("""
    INSERT INTO direcciones (user_id, street, suite, city, zipcode, lat, lng)
    VALUES (?, ?, ?, ?, ?, ?, ?)
    """, (u['id'], addr['street'], addr['suite'], addr['city'], addr['zipcode'], geo['lat'], geo['lng']))
    
    # Insertar en empresas
    comp = u['company']
    cursor.execute("""
    INSERT INTO empresas (user_id, name, catchPhrase, bs)
    VALUES (?, ?, ?, ?)
    """, (u['id'], comp['name'], comp['catchPhrase'], comp['bs']))
```

```python
conn.commit()
conn.close()
```
